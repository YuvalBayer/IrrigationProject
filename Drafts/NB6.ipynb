```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
import time
import Functions as fun
import seaborn as sns
sns.set_context("paper",font_scale=2)
```

```python
lines = fun.import_lines()
main_line = fun.import_main_line()
fun.plot_map(False, False) # Plotting map
plt.grid(b=False)

for line in lines: # Plotting drip lines
    line = line.reshape(2,2)
    plt.plot(line[:,0],line[:,1],c='blue')

plt.plot(main_line[:,0],main_line[:,1],c='blue',lw=3) # Plotting main line
```

```python
fig, axis = plt.subplots(figsize=(20, 13),dpi=200)
indices = np.arange(1,len(lines)+1)
for i,line in zip(indices,lines): # Plotting drip lines
    line = line.reshape(2,2)
    D = fun.calculate_distance(line)
    
    if (i%5 == 0):
        plt.plot([line[0,0],line[0,0]+D],[line[0,1],line[0,1]],c='tab:red')
        plt.scatter(np.linspace(line[0,0],line[0,0]+D,7)[1:-1],
                 np.repeat(line[0,1],7)[1:-1],
                 s = 40,
                 marker='x',
                 c='tab:red')
        plt.text(line[0,0]+D+10,line[0,1],str(i),c='tab:red')
        plt.plot([line[0,0]+D,line[0,0]+D],[line[0,1]+1.5,line[0,1]-1.5],c='tab:red')
        plt.plot([line[0,0]+D-1.5,line[0,0]+D],[line[0,1]-1.5,line[0,1]-1.5],c='tab:red')
        plt.plot([line[0,0]+D-1.5,line[0,0]+D],[line[0,1]+1.5,line[0,1]+1.5],c='tab:red')
    elif i > 40:
        plt.plot([line[0,0],line[0,0]+D],[line[0,1],line[0,1]],c='k')
        plt.scatter(np.linspace(line[0,0],line[0,0]+D,4)[1:-1],
                 np.repeat(line[0,1],4)[1:-1],
                 s = 40,
                 marker='x',
                 c='k')
        plt.plot([line[0,0]+D,line[0,0]+D],[line[0,1]+1.5,line[0,1]-1.5],c='k')
        plt.plot([line[0,0]+D-1.5,line[0,0]+D],[line[0,1]-1.5,line[0,1]-1.5],c='k')
        plt.plot([line[0,0]+D-1.5,line[0,0]+D],[line[0,1]+1.5,line[0,1]+1.5],c='k')
    else:
        plt.plot([line[0,0],line[0,0]+D],[line[0,1],line[0,1]],c='k')
        plt.scatter(np.linspace(line[0,0],line[0,0]+D,7)[1:-1],
                 np.repeat(line[0,1],7)[1:-1],
                 s = 40,
                 marker='x',
                 c='k')
        plt.plot([line[0,0]+D,line[0,0]+D],[line[0,1]+1.5,line[0,1]-1.5],c='k')
        plt.plot([line[0,0]+D-1.5,line[0,0]+D],[line[0,1]-1.5,line[0,1]-1.5],c='k')
        plt.plot([line[0,0]+D-1.5,line[0,0]+D],[line[0,1]+1.5,line[0,1]+1.5],c='k')
    
axis.set_aspect('equal', 'box')
plt.ylim(794100, 794450)
plt.xlim(259890, 260050)
plt.axis('off')
plt.savefig('Figures/sketch.png',bbox_inches='tight')
```
